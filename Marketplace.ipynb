# ШАГ 1

Задача: максимизировать прибыль Ozon.
Продавец поставил нам склад Х кол-во товара. Нам необходимо развести Х кол-во товара по n складам (предположительно один склад на город). В этих n городах есть спрос $y = (y_{1}, ..., y_{n})$ на товар. Необходимо так развести товары, чтобы максимально покрыть спрос, то есть $x_{1}\approx y_{1}, ..., x_{n}\approx y_{n}$.

Будем исходить из того, что у нас имеется один вид товара и n складов(городов).

# ШАГ 2

$c_{i}$ - стоимость перевозки 1 ед. товара в город i

$m_{i}$ - максимальный кол-во товаров для перевозки в город i

$p_{i}$ - комиссия маркетплейса с продажи 1 ед. товара в городе i

$x_{i}$ - кол-во товаров отправленных в город i

$s_{i}$ - цена за хранение 1 ед. товара на складе в городе i

$\mu_{i}$ - ожидаемый спрос на товар в городе i

$y_{i}$ - реальный спрос на товар в городе i

$F_{i}$ - функция распределения спроса в городе i

$$max  \quad-c^Tx+p^Tmin(x_{i},\mu_{i})-(p-s)^Tmax(x_{i}−\mu_{i},0)−(p-c)^Tmax(0,\mu_{i}-x_{i})\quad(1)$$

$$\sum x_{i}=X \quad (2)$$
$$0\leq x_{i}\leq m_{i} \quad (3)$$
$$0 \leq \mu_{i} \quad (4)$$
$$0\leq c_{i}\leq p_{i}\quad (5)$$
$$0\leq s_{i}\leq p_{i}\quad (6)$$

Прибыль маркетплейса (1) складывается из затратов на перевозку товаров в разные города и дохода, получаемого от продажи товаров (комиссии). Если кол-во товаров превышает реальный спрос, то мы теряем прибыль равную комисси минус доход от хранения товаров (p-s), так мы могли бы продать эти товары там, где на них есть спрос и получить на (p-s) больше. Если спрос превышает кол-во доступных товаров, то мы теряем прибыль (p-c), которую могли бы получить, если бы доставили больше товаров. Товары делятся и развозяться по n складам(городам) и в сумме эти товары равны кол-ву привезенных товаров поставщиком(2). Транспорт, на котором перевозят товары, обладает максимальной грузоподемностью (пропускной способностью), то есть за раз нельзя перевезти товаров больше, чем максимальная вместимость транспорта(3). Ожидаемый спрос больше или равен нулю (4). Доходы от комисии, которую мы взымаем, должна превышать расходы на транспортирорвку товаров(5), иначе мы будем работать в убыток, и должны превышать цену за хранение(6), иначе нам будет выгоднее просто хранить товар на складе, а не продавать их. 

# ШАГ 3

Мы не можем точно знать сколько товаров у нас купят, но мы можем сравнить потери, которые мы понесем в случаях, когда спрос превысит кол-во доступных товаров и когда у на складе останется излишел товара. Минимизируя эти потери мы можем найти оптимальное кол-во необходимого товара $x^\star$.

$l_{0} = c - s$ - убытки, которые мы терпим, если не реализовали весь товар (на 1 ед. товара) 

$l_{u} = p - c$ - прибыль, которую мы теряем, если не удовлетворяем спрос (на 1 ед. товара)

$y$ - реальный спрос на товар

Выпишем функцию минимизации расходов:
$$min\quad z=E[l_{0}max(0,x-y)+l_{u}max(0,y-x)]$$
$$=\int_{y=0}^{\infty}[l_{0}max(0,x-y)+l_{u}max(0,y-x)]f(y)dy$$
$$=\int_{y=0}^{x}[l_{0}(x-y)]f(y)dy+\int_{y=x}^{\infty}[l_{u}(y-x)]f(y)dy$$
$$=l_{0}\int_{y=0}^{x}(x-y)f(y)dy+l_{u}\int_{y=x}^{\infty}(y-x)f(y)dy$$
$$=l_{0}[x\int_{y=0}^{x}f(y)dy-\int_{y=0}^{x}yf(y)dy]+l_{u}[\int_{y=0}^{x}yf(y)dy-x\int_{y=0}^{x}f(y)dy]$$
$$=l_{0}[xF(x)-\int_{y=0}^{x}yf(y)dy]+l_{u}[\int_{y=0}^{x}yf(y)dy-x(1-F(x))]$$
$$z = l_{0}[xF(x)-\int_{y=0}^{x}yf(y)dy]+l_{u}[\int_{y=0}^{x}yf(y)dy-x(1-F(x))]$$
$$dz/dx = l_{0}[F(x)+xf(x)-xf(x)]+l_{u}[-xf(x)-(1-F(x)+xf(x))]$$
$$=l_{0}F(x)+l_{u}F(x)$$

$$l_{0}F(x^\star)+l_{u}F(x^\star)=0, $$ 
$x^\star$ - кол-во товаров, которое минимизирует ожидаемые потери

$$F(x^\star) = P(y\leq x^\star) = \frac{l_{u}}{l_{u}+l_{0}}\quad (1)$$
$$1 - F(x^\star) = P(y\geq x^\star) = \frac{l_{0}}{l_{u}+l_{0}}\quad (2)$$

(1) - вероятность того, что спрос будет меньше, или равен запасу товаров

(2) - вероятность того, что спрос будет больше, или равен запасу товаров

Предположим, что спрос на товары распределен равномерно на промежутке $(a,b)$, тогда $x^\star$ можно найти из следующего уравнения:

$$max(F(x^\star), 1-F(x^\star))=\frac{x^\star-a}{b-a}$$

# ШАГ 4

**Two stage problem**

1 stage:
$$max  -c^Tx^\star+E[Q(x^\star,\mu_{i})]$$
$$\sum x^\star_{i}=X$$
$$0\leq x^\star_{i}\leq m_{i}$$
$$ 0\leq c_{i}$$
$$0 \leq \mu_{i}$$

2 stage: 

$y_{i}$ - реальный спрос на товар в городе i

$$max  \quad p^Tmin(x^\star_{i},y_{i})-P(x^\star_{i}-y_{i})  \quad (1)$$

$$P(Z)=(p-s)^T(Z^+)+(p-c)^T(Z^-)$$
$$Z^+=max(z,0)$$
$$Z^-=max(-z,0)$$
$$z=x^\star_{i}-y_{i}$$
$$x^\star_{i},y_{i} \geq 0$$
$$0 \leq s_{i}\leq p_{i}$$
$$ 0\leq c_{i}\leq p_{i}$$

(1) - мы получаем прибыль от кол-ва товара, который был продан (если спрос меньше кол-ва товаров, то мы можем продать $y_{i}$, если товаров больше или равно - $x^\star_{i}$), и штрафуем себя за то, что у нас остается излишек товара или неудовлетворенный спрос (возможен только один вариант).


**Multistage problem**
$$max -c^Tx^\star_{i,t}+E[p^Tmin(x^\star_{i,t},y_{i,t})-P(x^\star_{i,t}-y_{i,t})+Q_{t+1}(x^\star_{i,t+1}-(x^\star_{i,t}-y_{i,t}), y_{i,t+1})] \quad (2)$$


$$Q_{t+1}(u_{i,t}, y_{i,t+1})=-c^Tu_{i,t}+E[p^Tmin(u_{i,t},y_{i,t+1})-P(u_{i,t}-y_{i,t+1})+Q_{t+2}(x^\star_{i,t+2}-(x^\star_{i,t+1}-y_{i,t+1}))]$$
$$P(Z)=(p-s)^T(Z^+)+(p-c)^T(Z^-)$$
$$Z^+=max(z_{i},0)$$
$$Z^-=max(-z_{i},0)$$
$$z_{i}=x^\star_{i}-y_{i}$$
$$0\leq x_{i}\leq m_{i}$$
$$x_{i},y_{i} \geq 0$$
$$0 \leq s_{i}\leq p_{i}$$
$$ 0\leq c_{i}\leq p_{i}$$

(2) - перевозим оптимальное кол-во товара $x^\star_{i,t}$ на склад i в момент времени t, продаем товары и в зависимости от спроса считаем штраф. На следующем шаге наше оптиматьное кол-во товара $x^\star_{i,t+1}$ необходимо скоректировать на разницу доступных товаров и спроса в момент времени t. Если товаров было больше $([x^\star_{i,t}-y_{i,t}]^+)$, чем спрос на них, то сократить объем поставляемых товаров на $(x^\star_{i,t}-y_{i,t})$, так они еще остались на складе i. Если спрос был больше $([x^\star_{i,t}-y_{i,t}]^-)$, то увеличить объем поставляемых товаров на $(-x^\star_{i,t}+y_{i,t})$, так как спрос не был удовлетворен в момент времени t и мы ожидаем, что неудовлетворенный спрос перейдет в момент времени t+1.
